Create DataFrame

```python
import pandas as pd

# create a data frame
print("# Show entire DataFrame")
data_cols = ["Name", "Class", "Score1"]
data_rows = [["Ellan", "CS", 87], ["Andrew", "Art", 98], ["Boby", "Acting", 83], ["Aqira", "Art", 78]]
df = pd.DataFrame(data=data_rows, columns=data_cols)
df
```

```python
# Create a new colum
print("# Create a new column")
df["Score2"] = 84, 97, 67, 72
df["Score3"] = 56, 86, 98, 70
df["Total"] = df["Score1"] + df["Score2"] + df["Score3"]
# copy a DataFrame
df_copy = df.copy()
df
```
