Import and read CSV file

```python
# importing the depencies
import pandas as pd
```

```python
# read CSV file
df_csv = pd.read_csv("Files/people.csv")
# the dataframe will be truncated by 5 from the top and 5 from the bottom
df_csv
```

```python
# print entire dataframe
print(df_csv.to_string())
```

```python
# print 5 data frames from top
df_csv.head(5)
```

```python
# print 5 data frames from bottom
df_csv.tail(5)
```

```python
# find specific gender (Male)
male = df_csv[df_csv["Sex"] == "Male"]
male.head(5)
```

```python
# number of data (male)
print(f"DataFrame size = {male.shape}")
print(f"Number of Male = {male.shape[0]}")
```

```python
# find specific gender (Female)
female = df_csv[df_csv["Sex"] ==  "Female"]
female.head()
```

```python
# number od data (female)
print(f"DataFrame size = {female.shape}")
print(f"Number of Female = {female.shape[0]}")
```

```python
# find poeple who born above 1999
teenagers = df_csv[df_csv["Date of birth"] > "1999-12-31"]
teenagers.head(5)
```

```python
# number of data (date of birth > 1999-12-31)
print(f"DataFrame size = {teenagers.shape}")
print(f"Number of People born above 1999-12-31 = {teenagers.shape[0]}")
```

Import and read Json file

```python
# read a json file
df_json = pd.read_json("Files/people.json")
# the dataframe will be truncated by 5 from the top and 5 from the bottom
df_json
```

```python
# show entire dataframe
print(df_json.to_string())
```

```python
# show specific data by language
language_uyghur = df_json[df_json["language"] == "Uyghur"]
language_uyghur.head(5)
```

```python
# number of data (uyghur language)
print(f"DataFrame size = {language_uyghur.shape}")
print(f"Number of Uyghur language users = {language_uyghur.shape[0]}")
```

```python
# show specific data by version
above_nine = df_json[df_json["version"] > 9.0]
above_nine.head(5)
```

```python
# number of data (version)
print(f"DataFrame size {above_nine.shape}")
print(f"Version above 9.0 = {above_nine.shape[0]}")
```
