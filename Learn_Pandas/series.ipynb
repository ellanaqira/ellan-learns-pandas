Series (1-Dimensional label array)

```python
import pandas as pd
```

- create a series from list

```python
# create a series from a list
name = ['Boby', 'Andrew', 'Rudy']
print(pd.Series(name), '\n')

score = [89, 93, 82]
print(pd.Series(score), '\n')

temp = [32.4 , 30.5, 33.3]
print(pd.Series(temp), '\n')

boolean = [True, False, True, False]
print(pd.Series(boolean), '\n')
```

```python
# custom index
print("# Custom index")

num_series = pd.Series(score, index=['a', 'b', 'c'])
print(num_series)

# accessing specific value
print(f"\nThe value of 'b' is = {num_series.loc['b']}")

print("\nValue that less than 90 :")
print(num_series[num_series < 90])
```

```python
# change series content
print("# Series before")
print(num_series)

print("\n# series after get changed\n-with loc")
# with loc
num_series.loc['b'] = 100
print(num_series)

print("\n-with iloc")
# with iloc (locate by int)
num_series.iloc[2] = 200
print(num_series)
```

- create a series from dictionary

```python
# create a series from a list
print("# Calories Track")
calories = {"Day1" : 1290, "Day2": 1306, "Day3" : 1248, "Day4" : 1809, "Day5" : 1590}
cals_df = pd.Series(calories)
print(cals_df, '\n')

# accessing specific value
print("# the day you eat >= 1500 calories")
print(cals_df[cals_df >= 1500])

print("\nday with the highest number of calories")
print(f"Day 4 : {cals_df.loc['Day4']}")
print(f"{cals_df.iloc[3] - 1500} calories excess :)")
```
